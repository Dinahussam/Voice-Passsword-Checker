```python
import os
path=r"D:\Projects\Voice recognition\DSP Task 3\Team 2"
X= np.zeros(256)
Words=np.array([])
Members=np.array([])
for word in os.listdir(path):
    if word=="Wrong":
        continue
    for member in os.listdir(os.path.join(path,word)):
        for voice in os.listdir(os.path.join(path,word,member)):
            samples, sr= librosa.load(os.path.join(path,word,member,voice))
            test_stft = librosa.stft(samples, n_fft=2048)
            test_stft_mag = np.abs(test_stft)
            h1 = imghash.phash(test_stft_mag)
            X=np.vstack((X,h1.hash.flatten()))
            Words=np.append(Words,word )
            Members=np.append(Members,member )
            
```

```python
Data=pd.DataFrame(X).drop(0)
Data["Words"]=Words
Data["Members"]=Members
```

```python
Data["Total"]=Data["Words"]+" "+Data["Members"]
Data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test= train_test_split(Data.drop(["Words","Members","Total"],axis=1),Data["Total"],stratify=Data["Total"])
```

```python
from sklearn.ensemble import RandomForestClassifier
clf= RandomForestClassifier(n_estimators=150,n_jobs=-1)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
np.mean(y_pred==y_test)
```

```python
from sklearn.metrics import f1_score
y_pred_train=clf.predict(X_train)
f1_score(y_true=y_train,y_pred=y_pred_train,average="weighted")
```
